```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.transform as st
```

```python
# Load Lung images
filename = 'lung_img-17756198_53155374.jpg' # The lung image path
lung_img = cv.imread(filename)
plt.imshow(lung_img)
plt.show()
```

```python
# Load Original image
orginal_img = cv.imread('01ce86a4-1dfc96d7-5504e0a6-6de4a05f-4149c647.jpg')
orginal_img = st.resize(orginal_img, (512, 512))
plt.imshow(orginal_img)
plt.show()
```

```python
# Load images
bg     = orginal_img                                                               
obj = lung_img                                                      
mask   = lung_img                                                           

# Zero background where we want to overlay
# Here we use the lung image as a mask
bg[mask>0]=0                                                                              

# Add object to zeroed out space
bg += obj*(mask>0)     

plt.imshow(bg)
plt.show()
```
